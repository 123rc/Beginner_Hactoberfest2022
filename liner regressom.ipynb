```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets,linear_model
from sklearn.metrics import mean_squared_error 
```

```python
diabetes =datasets.load_diabetes()
```

```python
print(diabetes.DESCR)
```

```python
print(diabetes.keys())
```

```python
d_x=diabetes.data[:,np.newaxis,2]
print(d_x)
```

```python
print(diabetes.frame)
```

```python
d_x_train=d_x[:-50]
```

```python
d_x_test=d_x[-20:]
```

```python
d_y_train=diabetes.target[:-50]
d_y_test=diabetes.target[-20:]
```

```python
model=linear_model.LinearRegression()
```

```python
model.fit(d_x_train,d_y_train)
```

```python
d_y_predicted=model.predict(d_x_test,)
```

```python
print(mean_squared_error(d_y_test,d_y_predicted))
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
plt.scatter(d_x_test,d_y_test)
plt.plot(d_x_test,d_y_predicted)
plt.show()
```
